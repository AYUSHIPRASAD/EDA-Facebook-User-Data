## Data Preprocessing

```python
df.drop_duplicates(inplace=True)
```

```python
df = df.dropna()
```

```python
df_old =df[df['age'] > 90]
df_old.describe()
```

```python
df_old[df_old['friend_count'] != 0]
```

```python
print(min(df["age"]))

print(max(df["age"]))
```

```python
age_grp=['10-20','21-30','31-40','41-50','51-60','61-70','71-80','81-90','91-100','101-110','111-120']
df['age_group'] = pd.cut(df.age,bins=np.arange(10,121,10),labels=age_grp,right=True)
df.head()
```

```python
df.groupby(['age_group'])['age_group'].count()
```

```python
df["tenure"] = df["tenure"].fillna(0)
```

```python
df["tenure"] = df["tenure"].apply(np.int64)
```

```python
df['tenure_group'] = pd.cut(df['tenure'],
                      [0,30,90,180,365,730,1803,3650],
                      labels=['<1M','1M-3M','3M-6M','6M-1Y','1Y-2Y','2Y-5Y','5Y+'], include_lowest=True)
```

## Exploratory Data Analysis

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,5))
sns.countplot(x='age_group',data=df,palette="rocket" )
plt.title("User count by Age-Group",fontsize=20,fontweight="bold")
```

####  Most user counts are in Age group between 21 yrs to 30 Yrs.

```python
sns.set_style('whitegrid')
plt.figure(figsize=(8,5))

sns.countplot(x='gender', data = df, palette="rocket")
plt.title("User count by Gender",fontsize=20,fontweight="bold")
plt.ylabel("User Count")
plt.xlabel("Gender")
```

#### Most users are Male

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='age_group',y='friend_count',data=df,color='purple')
plt.title("Friend Count by Gender Group",fontsize=20,fontweight="bold")
plt.ylabel("Friend Count",fontsize=15 )
plt.xlabel("Gender Group", fontsize=15)
```

#### Gender Group 10-20 yrs have most Friend Count

```python
plt.figure(figsize=(12,5))
sns.countplot('tenure_group', data = df, palette="gnuplot")
plt.title("User count by Tenure",fontsize=20,fontweight="bold")
plt.ylabel("User Count", fontsize=15)
plt.xlabel("Tenure of Users",  fontsize=15)
```

#### Most user counts are in Tenure group between 1 yrs to 2 Yrs

```python
df["mbl_activity"] = df["mobile_likes"] + df["mobile_likes_received"]
df["web_activity"] = df["www_likes"] + df["www_likes_received"]
df.tail()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x='age_group', data=df, hue='gender', palette="rocket")
plt.title("Count plot for Age Group and Gender",fontsize=20,fontweight="bold")
plt.ylabel("User Count",fontsize=15)
plt.xlabel("Age group of Users",fontsize=15)
```

#### Most user counts are in Age group between 21 yrs to 30 Yrs.

#### Most Male and female user counts are in in Age group between 21 yrs to 30 Yrs.

```python
plt.figure(figsize=(12,10))
sns.countplot(x='tenure_group', data=df, hue='gender', palette="rocket")
plt.title("Count plot for Tenure Group and Gender",fontsize=20,fontweight="bold")
plt.ylabel("User Count",fontsize=15)
plt.xlabel("Age group of Users",fontsize=15)
```

#### Most user counts are in Tenure group between 1 yrs to 2 Yrs.

#### Most male and female user counts are in Tenure group between 1 yrs to 2 Yrs.

```python
plt.figure(figsize=(12,5))
sns.kdeplot(df['dob_year'], color='purple',fill=True)
plt.title("Kernel Distribution Estimation Plot for Birth Year and Density",fontsize=20,fontweight="bold")
plt.ylabel("Year of Birth", fontsize =15)
plt.xlabel("Users Density", fontsize =15)
```

#### Maximum users are born between 1980-2000.

```python
plt.figure(figsize=(12,5))
sns.pointplot(x='age_group',y='mbl_activity',data=df ,hue='gender',palette='magma')
plt.title("Point Plot for Mobile Activity and Gender",fontsize=20,fontweight="bold")
plt.ylabel("Mobile Activity",fontsize=15)
plt.xlabel("Gender",fontsize=15)
```

#### Female users have greater Mobile Activity than male users.

```python
plt.figure(figsize=(12,5))
sns.pointplot(x='age_group',y='web_activity',data=df ,hue='gender',palette='magma')
plt.title("Point Plot for Web Activity and Gender",fontsize=20,fontweight="bold")
plt.ylabel("Web Activity",fontsize=15)
plt.xlabel("Gender",fontsize=15)
```

#### Female users have greater Web Activity than male users.

```python
df["gender"].value_counts().plot(kind='pie',fontsize=14, autopct='%3.1f%%', wedgeprops=dict(width=0.20), 
                                       shadow=True, startangle=160, cmap ='Spectral', figsize=(10,10),  legend=True)

plt.ylabel('Gender', fontsize=15)
plt.title('%age of Male & Female Distribution',fontsize=20,fontweight="bold")
```

#### There are more Male users than Female users.

```python
space = np.ones(11)/11
df['age_group'].value_counts().plot(kind='pie',explode=space,  fontsize=10, autopct='%3.1f%%', wedgeprops=dict(width=0.20), 
                                       shadow=True, startangle=160, figsize=(15,15), cmap='Spectral', legend=True)
plt.ylabel('Gender', fontsize=15)
plt.title('%age of Age Group Distribution',fontsize=20,fontweight="bold")
```

#### Highest number of users are aged between 21 to 20 years.

#### Around 54% users are less then 30 years old.

```python
df.groupby(['gender', 'age_group'])['friend_count'].mean().plot(kind='pie', autopct='%3.1f%%', wedgeprops=dict(width=0.15), 
                                                                figsize=(16,18), fontsize=11, 
                                                                 startangle=20, shadow=True, cmap='Spectral')
plt.ylabel('Friend Count', fontsize=15)
plt.title('%age of Friends count wrt Gender & Age Group',fontsize=20,fontweight="bold")
```

#### Top 3 categories for number of friend counts are Females between 10 to 20, Females between 101 to 110 & Males between 110 & 110.

```python
df.groupby(['gender', 'age_group'])['likes'].mean().plot(kind='pie', autopct='%3.1f%%', wedgeprops=dict(width=0.15), 
                                                                figsize=(15,15), fontsize=11, 
                                                                 startangle=20, shadow=True, cmap='Spectral')
plt.ylabel('Likes', fontsize=15)
plt.title('%age of Likes wrt Gender & Age Group',fontsize=20,fontweight="bold")
```

#### Top 3 categories for number of likes are Females between 10 to 20, Females between 41 to 50 & Females between 51 & 60.
